```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import os
import pandas as pd 
from io import StringIO
import numpy as np
from bs4 import BeautifulSoup
```

```python
chromedriver_path = "chromedriver.exe"

# Inicializar el servicio de ChromeDriver
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)
```

```python
URL = "https://listado.mercadolibre.com.ec/laptop#D[A:laptop]"
```

```python
driver.get(URL)
```

```python
def get_html(url, selector, sleep_time=20):
    driver.get(url)
    time.sleep(sleep_time)
    element = driver.find_element(By.CSS_SELECTOR, selector) #Aprovechamos el find_element para obtener el elemento de css que nos devuelve el html
    return element.get_attribute("outerHTML")
```

```python
html_content = get_html(URL, "#root-app > div > div.ui-search-main.ui-search-main--without-header.ui-search-main--only-products.ui-search-main--4x")
```

```python
soup = BeautifulSoup(html_content)
```

```python
soup.find_all("h3")
```

```python
titlelinks = soup.find_all("a", class_="poly-component__title")

for link in titlelinks:
    print("Título:", link.get_text(strip=True))
    print("URL:", link["href"])
    print("-" * 60)
```

```python
prices = soup.select(
    "span.andes-money-amount.andes-money-amount--cents-superscript"
)

for price_tag in prices:
    print(price_tag.get_text(strip=True))
```

```python
products = []

for product in soup.find_all("div", class_="poly-card__content"):
    
    title_tag = product.find("a", class_="poly-component__title")
    price_container = product.find("div", class_="poly-price__current")
    
    if title_tag and price_container:
        
        title = title_tag.get_text(strip=True)
        url = title_tag["href"] 
        
        price_whole = price_container.find("span", class_="andes-money-amount__fraction")
        price_cents = price_container.find("span", class_="andes-money-amount__cents")
        
        if price_whole:
            price = price_whole.get_text(strip=True)
            if price_cents:
                price += "." + price_cents.get_text(strip=True)
            
            products.append({
                "title": title,
                "price": float(price),
                "url": url 
            })
```

```python
df = pd.DataFrame(products)
```
